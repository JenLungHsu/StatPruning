```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# 資料整理
data = {
    "Model": [
        "CLR-RNF", "DECORE_1", "DECORE_2", "DECORE_3", "EPruner", "ESNB", "FRPP",
        "GAL_1", "GAL_2", "HRank_1", "HRank_2", "HRank_3", "L1_1", "L1_2", "NISP",
        "Ours(KS)", "Ours(MW)", "PLS", "TMI-GKP", "Xu et al_1", "Xu et al_2", "Xu et al_3", "Zhao et al"
    ],
    "DeltaAcc": [
        0.14, 0.38, 0.0, -0.79, 0.12, 0.07, 0.03,
        0.09, -0.76, 0.73, -0.14, -0.85, 0.02, -0.23, -0.18,
        1.13, 1.06, -0.06, 0.64, 0.29, -0.13, -0.05, -0.25
    ],
    "ParamsPR": [
        69.1, 35.7, 64.8, 79.6, 76.3, 25.19, 25.19,
        4.1, 44.8, 39.4, 59.2, 68.7, 2.3, 32.4, 43.25,
        60.47, 54.07, 34.77, 43.52, 8.7, 55.5, 69.4, 41.27
    ]
}

df = pd.DataFrame(data)
df["BaseModel"] = df["Model"].apply(lambda x: x.split("_")[0].replace(" ", ""))
df["DisplayName"] = df["BaseModel"].replace({
    "Xuetal": "Xu et al.",
    "Zhaoetal": "Zhao et al.",
    "Ours(MW)": "Ours (MW)",
    "Ours(KS)": "Ours (KS)"
})

# 顏色設定
unique_models = df["BaseModel"].unique()
colors = plt.cm.tab20(np.linspace(0, 1, len(unique_models)))
model_color_map = dict(zip(unique_models, colors))

# 可放在下方的標籤
place_rightbelow = {"Zhao et al.", "L1", "PLS", "Ours (MW)"}
place_leftup = {"ESNB", "FRPP"}

# 畫圖
fig, ax = plt.subplots(figsize=(11, 7))
labeled_models = set()
legend_handles = {}

for model in unique_models:
    sub_df = df[df["BaseModel"] == model].sort_values("ParamsPR")
    x = sub_df["ParamsPR"]
    y = sub_df["DeltaAcc"]
    color = "red" if "Ours" in model else model_color_map[model]
    display_name = df[df["BaseModel"] == model]["DisplayName"].values[0]

    if "Ours" in model:
        ax.scatter(x, y, color=color, marker="*", s=200, label=display_name)
    else:
        ax.plot(x, y, marker="o", linestyle="-", color=color, label=display_name)

    if display_name not in labeled_models:
        if display_name in place_rightbelow:
            dx, dy = (1.0, -0.065)
        elif display_name in place_leftup:
            dx, dy = (-5.0, -0.01)
        else:
            dx, dy = (0.0, 0.05)
        fontweight = 'bold' if "Ours" in display_name else 'normal'
        ax.text(x.values[0] + dx, y.values[0] + dy, display_name, fontsize=9, color=color, fontweight=fontweight)
        labeled_models.add(display_name)
        legend_handles[display_name] = plt.Line2D([0], [0], color=color, marker='o', linestyle='-')

# Δ Accuracy = 0 baseline
ax.axhline(0, linestyle="--", color="gray", linewidth=1)

# 標籤與標題
ax.set_xlabel("Parameter Pruning Rate (%)", fontsize=12)
ax.set_ylabel("Δ Accuracy (%)", fontsize=12)
ax.set_xticks([0, 20, 40, 60, 80, 100])
ax.set_title("Δ Accuracy vs. Parameter Pruning Rate on ResNet110 using CIFAR-10", fontsize=14, pad=20)

# 圖例：將 Ours 系列放最後
ordered_legend = sorted(legend_handles.items(), key=lambda x: 0 if "Ours" not in x[0] else 1)
ax.legend([h for _, h in ordered_legend], [l for l, _ in ordered_legend],
          loc="upper right", fontsize=10, title="Method", title_fontsize=11)

ax.grid(True)
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# 你的資料
methods = [
    ("ResNet56", 0.00, 93.40),
    ("EMD-0.025", 15.29, 94.19),
    ("Ours-MW", 21.18, 94.24),
    ("Ours-KS", 25.88, 94.64),
    ("EMD-0.05", 31.76, 93.76),
    ("EMD-0.075", 77.65, 91.76),
    ("EMD-0.1", 78.82, 91.36)
]

# 分類資料
emd_x, emd_y, emd_labels = [], [], []
ours_x, ours_y = [], []

for name, px, acc in methods:
    if "Ours" in name:
        ours_x.append(px)
        ours_y.append(acc)
    else:
        emd_x.append(px)
        emd_y.append(acc)
        emd_labels.append(name)

# 開始繪圖
plt.figure(figsize=(10,6))

# EMD 與 ResNet56 折線圖與圓點
plt.plot(emd_x, emd_y, color='orange', label='Test Accuracy')
plt.scatter(emd_x, emd_y, color='orange', s=30, marker='o', zorder=3)
plt.axhline(y=93.40, linestyle='dotted', color='orange', label='Baseline Accuracy')
plt.title("Performance vs. Params Pruning Ratio", pad=20)

# EMD 標籤，特別處理 EMD-0.025 往左避免重疊
for i in range(len(emd_x)):
    offset_x = -4 if emd_labels[i] == "EMD-0.025" else 0.5
    plt.text(emd_x[i]+offset_x, emd_y[i], emd_labels[i], fontsize=10)

# Ours 星星與標註 (固定乾淨標籤避免空格問題)
ours_labels_fixed = ["Ours-MW", "Ours-KS"]
for i in range(len(ours_x)):
    plt.scatter(ours_x[i], ours_y[i], color='orange', s=100, marker='*', zorder=4)
    plt.text(ours_x[i]+0.5, ours_y[i], ours_labels_fixed[i] , fontsize=10, weight = 'bold')

plt.xlabel("Params Pruning Ratio (%)")
plt.ylabel("Accuracy (%)")
plt.grid(True)
plt.legend()
plt.show()
```
